```python
import json
import matplotlib.pyplot as plt
import numpy as np
```

```python
with open("face.json", "r") as f:
    face = json.load(f)
```

```python
plt.rcParams["figure.figsize"] = (5,5)

xrr = []
yrr = []
zrr = []

for keypoint in face:
    xrr.append(keypoint[0])
    yrr.append(keypoint[1])
    zrr.append(keypoint[2])

xrr = np.array(xrr)
yrr = np.array(yrr)
zrr = np.array(zrr)

plt.scatter(xrr, yrr, s = 5)
plt.xlim([xrr.min(), xrr.max()])
plt.ylim([yrr.min(), yrr.max()])
plt.gca().invert_yaxis()
plt.gca().set_aspect('equal', adjustable='box')
```

```python
plt.rcParams["figure.figsize"] = (10,15)

fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.scatter3D(xrr, yrr, zrr, s = 15)
ax.elev = 90
ax.azim = 90
plt.xlim([xrr.min(), xrr.max()])
plt.ylim([yrr.min(), yrr.max()])
plt.gca().invert_yaxis()
```
